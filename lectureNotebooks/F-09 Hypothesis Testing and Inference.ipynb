## Student's t distribution

```python
%matplotlib inline
import random
from scipy import stats as ss
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi
import seaborn as sns
import scipy.stats

sns.set_style("darkgrid")

#make some points
x = np.linspace(-5,5, 100)

#Compute pdf on t and normal
y1 = scipy.stats.t.pdf(x, df=3)
y2 = scipy.stats.norm.pdf(x)
y3 = scipy.stats.t.pdf(x, df=10)

plt.plot(x,y1, label="t, n=3")
plt.plot(x,y3, label="t, n=10")
plt.plot(x,y2, label="Normal")

plt.legend()
```

### T-score Example

University of Reykjavik claims that its graduates earn **\$200 per hour on average **. After **sampling 9 graduates** we find their salaries have a mean of **\$180 with a standard deviation of \$30.**

What is the probability that the mean salary of graduates is no more than \$180, assuming what the school is claiming is true?

```python
sm = 180
m = 200
s = 30
n = 9

t = (sm-m)/(s/np.sqrt(float(n)))  # t-statistic for mean
print("The t-score is", t)
```

Knowing the t-statistic allows us to find the probability by looking at the probability distribution of t and summing the shaded area shown below:

```python
x = np.linspace(-5,5, 100)
y = scipy.stats.t.pdf(x, df=8)

plt.plot(x, y, label="t, n=8")
plt.fill_between(x,y,where=x<=-2) # fill in the area denoting the probability we're computing
plt.legend()
```

```python
# the probability can now be computed simply using the cdf
p = scipy.stats.t.cdf(t, df=n-1)
print("The probability that the mean salary of graduates is no more than ${0} is {1:.1%}".format(sm, p))
```

### Z-score Example
Cars traveling along Miklabraut were measured having average speed of 80 km/h, with a standard deviation of 8 km/h.
What is the probability that a randomly selected car is going 90 km/h or less?

```python
sm = 90
m = 80
d = 8

z = (sm-m)/d
print("The z-score is", z)
```

```python
x = np.linspace(40,120,100)
y = scipy.stats.norm.pdf(x, loc=80, scale=8)

plt.plot(x,y, label="mean=80, stdev=8")
plt.fill_between(x, y, where=x<=90) # fill in the area denoting the probability we're computing
plt.legend()
```

```python
# computing the probability can be done either by looking it up on a table (if you're over 40) or by computing it:
p = scipy.stats.norm.cdf(1.25)
print("The probability that a randomly selected car is going {0} km/h or less is {1:.1%}".format(sm, p))
```
