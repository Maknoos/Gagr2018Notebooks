#### Exercise 1
Write a function `even` that takes as input a list of integers and returns a copy of the list where all the odd numbers have been removed.

```python
def even(l):
    """Removes all odd numbers from the list l."""
    return [ i for i in l if i % 2 == 0]
```

Sample test:

```python
even([1, 4, 9, 16, 25, 36, 49, 64, 81, 100]) == [4, 16, 36, 64, 100]
```

#### Exercise 2
Write a function `word_reverse` that takes as input a string consisting of separated words and returns a new string where each word has been reversed.

```python
def word_reverse(s):
    """Reverses every word in the string s."""
    return ' '.join([ t[::-1] for t in s.split()])
```

Sample test case:

```python
word_reverse('this might be the best course i have ever taken') == 'siht thgim eb eht tseb esruoc i evah reve nekat'
```

#### Exercise 3

Write a function `frequency` that takes as input a string and returns the letter frequency of the string as a list of integers where the $i$-th value corresponds to the frequency of the $i$-th letter in the English alphabet.

```python
def frequency(s):
    """Computes the letter frequency of the string s."""
    l = [0]*26
    for c in s.lower():
        if c.isalpha():
            l[ord(c) - ord('a')] += 1
    return l
```

Sample test case:

```python
frequency('this might be the best course i have ever taken') == [2, 2, 1, 0, 8, 0, 1, 4, 3, 0, 1, 0, 1, 1, 1, 0, 0, 2, 3, 5, 1, 2, 0, 0, 0, 0]
```

#### Exercise 5
Write a function `pref_count` that takes as input a string and outputs the number of words from the [SOWPODS](data/sowpods.txt) dictionary that have the string as a prefix.

```python
def pref_count(s):
    """Returns the number of words in the SOWPOD dictionary that have s as a prefix."""
    with open('sowpods.txt', 'r') as f:
        words = f.read().splitlines()
        return len([w for w in words if w.startswith(s)])
```

```python
pref_count('AB') == 866
```

#### Exercise 7
Using `numpy`, write a function `rand2darr` that takes a single integer $n$ as input and returns a random $n \times n$ array and its minimum and maximum value. 

```python
def rand2darr(n):
    """Returns a random nxn array and its minimum and maximum value."""
    a = np.random.randint(0, 100, (n, n))
    return a, np.min(a), np.max(a)
```

```python
rand2darr(10)
```

#### Exercise 8
Using `numpy`, write a function `randarr` that takes as input two integers $n$ and $d$ and returns a random $n^d$ array and its mean value.

```python
def randarr(n, d):
    """Returns a random n^d array and its mean value."""
    arr = np.random.randint(0, n**2, tuple([n for i in range(d)]))
    return (arr, arr.mean())
```

```python
randarr(10, 3)
```

#### Exercise 10
Using `numpy`, write a function `colsort` that takes as input a $2$-dimensional array and an integer $c$ and returns the array sorted by the $c$-th column. 

```python
def colsort(m, c):
    """Returns the array 2-dimensional m sorted by the c-th column."""
    return m[np.argsort(m[:,c])]
```

```python
arr = randarr(10, 2)
sortedarr = colsort(arr[0], 4)
print(sortedarr)
```

#### Exercise 11
Using `numpy`, write a function `is_diagonal` that takes as input a $2$-dimensional array and checks if it is a diagonal matrix.

```python
def is_diagonal(m):
    """Checks if the 2-dimensional array m is a diagonal matrix."""
    return np.all(np.equal(np.diag(np.diag(m)), m))
```

```python
arr = np.diag(np.arange(0, 10))
print(arr)
is_diagonal(arr)
```
